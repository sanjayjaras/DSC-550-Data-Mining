```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
def calculate_distance_matrix(df):
    """
    calculate distance matrix for
    :param df: input dataset
    :return: distance matrix
    """
    n = df.shape[0]
    dist_matrix = np.zeros(shape=(12, 12))
    for i in range(n):
        for j in range(n):
            dist_matrix[i, j] = np.round(np.linalg.norm(df[i] - df[j]), 2)
    return dist_matrix


d = np.array([[4,10], [7,10], [4,8], [6,8], [3,4], [10,5], [2,2], [5,2], [9,3],[11,4],[12,3],[12,6]])

dist = calculate_distance_matrix(d)

df = pd.DataFrame(dist)
print(d)
print(df)

a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
ax = sns.heatmap(df, annot=True)
plt.show()
```
